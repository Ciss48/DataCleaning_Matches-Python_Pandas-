# Cleaning Data by Pandas 

Tổng quan dữ liệu: Dữ liệu thu được từ trang: https://www.espncricinfo.com/records/highest-career-batting-average-282910 (Xuất về file Exel). Khá thô và xấu, cần phải trải qua nhiều bước cleaning 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data.xlsx')
```

```python
df
```

```python
# Rename columns 
df = df.rename(columns={'Mat': 'matches', 
                   'Inns':'innings_batted',
                   'NO': 'not_outs',
                   'HS': 'highest_inns_score',
                   'Ave': 'batting_average',
                   'BF': 'ball_faced',
                   'SR': 'batting average'}, inplace=True)
```

```python
df
```

```python
# Check Null Values
```

```python
df.isnull().sum()
```

```python
df[df['batting average'].isnull() == True]
```

```python
# Đổi dữ liệu null trên thành 0 
df['batting average'] = df['batting average'].fillna(0)
df['ball_faced'] =df['ball_faced'].fillna(0)
```

```python
df.isnull().sum()
```

```python
# Drop duplicates
```

```python
#Check:
df.duplicated()
```

```python
df[df.duplicated() == True]
```

```python
df[df['Player'].isin(['AC Voges (AUS)','ED Weekes (WI)'])]
```

```python
df = df.drop_duplicates()
```

```python
df[df['Player'].isin(['AC Voges (AUS)','ED Weekes (WI)'])]
```

```python
# Split up Span into start_date and end_date
```

```python
df['start_date'] = df['Span'].str.split('-').str[0]
df['end_date'] = df['Span'].str.split('-').str[1]
```

```python
df
```

```python
#Drop Span column
df = df.drop(columns = 'Span')
df
```

```python
# Split up country from player and make it into a new column
df['Player'].str.split('(').str[1].str.split(')').str[0]
```

```python
df['country'] = df['Player'].str.split('(').str[1].str.split(')').str[0]
```

```python
df['Player'] = df['Player'].str.split('(').str[0]
```

```python
df
```

```python
#Change datatypes
```

```python
df.dtypes
```

```python
# Hàm xử lý ký tự đặc biệt
import re
def remove_special_chars(text):
    if isinstance(text, str):
        return re.sub(r'[^0-9]', '', text)
    else:
        return str(text)
```

```python
df['highest_inns_score']=df['highest_inns_score'].apply(remove_special_chars)
```

```python
df['highest_inns_score']=df['highest_inns_score'].astype('int')
```

```python
df['ball_faced']=df['ball_faced'].apply(remove_special_chars)
```

```python
df['ball_faced']=df['ball_faced'].astype('int')
```

```python
#start_date and end_date 
```

```python
df['start_date']=df['start_date'].astype('int')
df['end_date']=df['end_date'].astype('int')
```

```python
df.dtypes
```

```python
# Build a Career_length
```

```python
df['career_length'] = df['end_date'] - df['start_date']
df
```

```python
#Question 1: What is the average_length
df['career_length'].mean()
```

```python
#Question 2: Average highest_inns_score of player who has career length more than 10 years
df[df['career_length']>10]['ball_faced'].mean()
```

```python
#Question 3: Number of Players who played before 1960
df[df['start_date'] > 1960]['Player'].count()
```
